# Plotting data
Let's now try to plot some data and include it in this document as well as on the github repository.
First, load numpy for some basic array computations and functions and matplotlib for plotting.

```python
import numpy
import matplotlib.pyplot as plt
```

Now, let's create data for plotting.
In fact, we'll create three time series with random data.

```python
rng = numpy.arange(50)
rnd = numpy.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng
```

[yrs] is years and will be plotted in the x-axis.
The three series are created by the rng + rnd.

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')
ax.legend(loc='upper left')
ax.set_ylabel('Total debt')
ax.set_xlim(left=yrs[0], right=yrs[-1])
fig.tight_layout()
```

The denominations in matplotlib are a bit confusing.
A figure is the whole object that is created and can include several plots, subfigures, etc...
Axes are the individual plots, subfigures, etc... It is not connected to what we usually call axes (x-, and y-axis).
These are subclasses of axes.

Now, let's create a scatter plot and a histogram as subplots (axes) of the same figure. Starting with creating random data again.

```python
x = numpy.random.randint(low=1, high=11, size=50)
y = x + numpy.random.randint(1, 5, size=x.size)
data = numpy.column_stack((x, y))
```

Creating the figure first.
Then defining both subplots, ax1 and ax2.


```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,figsize=(8, 4))

ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')

ax2.hist(data, bins=numpy.arange(data.min(), data.max()),label=('x', 'y'))
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Frequencies of $x$ and $y$')
ax2.yaxis.tick_right()
```
